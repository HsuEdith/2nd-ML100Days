```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean_data = app_train['AMT_INCOME_TOTAL'].mean()
std_data = app_train['AMT_INCOME_TOTAL'].std()
print('AMT_INCOME_TOTAL-平均數 = ' + str(mean_data) + '\nAMT_INCOME_TOTAL-標準差 = ' + str(std_data))
```

```python
import pylab as pl #為將x軸字呈現轉角度import

original_arr = app_train['NAME_INCOME_TYPE'] #將 NAME_INCOME_TYPE 單獨出arr
distinct_arr = original_arr.drop_duplicates() #只取出差異
dsum_arr  = []
for ntype in distinct_arr:
    dsum_arr.append((original_arr == ntype).sum()) #計算差異總數

    
cnt_arr = len(distinct_arr) #計算差異個數
index_arr = np.arange(cnt_arr) #一定要轉型，不然xticks時，會出現錯誤

width = 0.5       # the width of the bars: can also be len(x) sequence，bar的寬度
ntype_bar = plt.bar(index_arr,dsum_arr,width)

plt.xlabel('NAME_INCOME_TYPE') #設定x軸的文字
pl.xticks(rotation=65)
plt.title('NAME_INCOME_TYPE Sum') #圖表表頭
plt.xticks(index_arr,distinct_arr) #設定x軸總共幾個點，點的文字
plt.grid(True) #格線

plt.show() #顯示圖表
```
